# Imprime los primeros libros con sus precios

```python
import requests
from bs4 import BeautifulSoup

# URL de la página principal de Books to Scrape
url = "http://books.toscrape.com/"

# Enviar solicitud GET a la página
response = requests.get(url)

# Verificar que la solicitud fue exitosa
if response.status_code == 200:
    # Parsear el contenido HTML de la página
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encontrar todos los bloques de libros
    books = soup.find_all('article', class_='product_pod')

    # Iterar sobre cada libro y extraer el título y el precio
    for book in books:
        title = book.h3.a['title']  # El título del libro está en el atributo 'title' del elemento 'a'
        price = book.find('p', class_='price_color').get_text()  # El precio está en el elemento 'p' con la clase 'price_color'
        print(f"Título: {title}\nPrecio: {price}\n")
else:
    print(f"Error al acceder a la página: {response.status_code}")
```

# WEBCRAPING
Solicita los titulos, precios y la categoria de cada libro de la pagina.
Los guarda en un archivo csv


```python
# URL base de la página de Books to Scrape
base_url = "http://books.toscrape.com/catalogue/page-{}.html"
category_url = "http://books.toscrape.com/catalogue/category/books_1/index.html"

# Lista para almacenar los datos de los libros
books_data = []

# Obtener la categoría de cada libro
def get_category(book_url):
    response = requests.get(book_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    category = soup.find('ul', class_='breadcrumb').find_all('a')[2].text.strip()
    return category

# Iterar sobre las páginas (hay 50 páginas en total)
for page in range(1, 51):
    # Formar la URL completa para la página actual
    url = base_url.format(page)

    # Enviar solicitud GET a la página
    response = requests.get(url)

    # Verificar que la solicitud fue exitosa
    if response.status_code == 200:
        # Parsear el contenido HTML de la página
        soup = BeautifulSoup(response.text, 'html.parser')

        # Encontrar todos los bloques de libros
        books = soup.find_all('article', class_='product_pod')

        # Iterar sobre cada libro y extraer el título, precio y categoría
        for book in books:
            title = book.h3.a['title'].encode('latin1').decode('utf-8')
            price = book.find('p', class_='price_color').get_text().strip()
            book_url = "http://books.toscrape.com/catalogue/" + book.h3.a['href']
            category = get_category(book_url)
            books_data.append([title, price, category])
    else:
        print(f"Error al acceder a la página: {response.status_code}")

# Guardar los datos en un archivo CSV
with open('books_titles_prices_categories.csv', 'w', newline='', encoding='utf-8-sig') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Price', 'Category'])
    writer.writerows(books_data)

print("Datos guardados en 'books_titles_prices_categories.csv'")
```

# Analisis Descriptivo

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo
df = pd.read_csv('libros_precios_categorias.csv')

# Convertir la columna 'Price' a valores numéricos
df['Price'] = df['Price'].str.replace('£', '').astype(float)

# Ver los primeros datos del archivo
df.head()
```

```python
# Estadísticas descriptivas de los precios
price_stats = df['Price'].describe()
print(price_stats)

# Media y mediana
mean_price = df['Price'].mean()
median_price = df['Price'].median()
mode_price = df['Price'].mode()[0]

print(f"Media del precio: £{mean_price:.2f}")
print(f"Mediana del precio: £{median_price:.2f}")
print(f"Moda del precio: £{mode_price:.2f}")

# Desviación estándar
std_dev = df['Price'].std()
print(f"Desviación estándar del precio: £{std_dev:.2f}")
```

```python
# Contar el número de libros en cada categoría
category_counts = df['Category'].value_counts()

# Convertir a un DataFrame para mejor visualización
category_counts_df = category_counts.reset_index()
category_counts_df.columns = ['Category', 'Number of Books']

# Mostrar la tabla
print(category_counts_df)
```

```python
# Histograma de la distribución de precios
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True, color='blue')
plt.title('Distribución de Precios de los Libros')
plt.xlabel('Precio (£)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Conteo de libros por categoría
category_counts = df['Category'].value_counts()

# Gráfico de barras de la distribución de categorías
plt.figure(figsize=(12, 8))
sns.barplot(x=category_counts.values, y=category_counts.index, palette='viridis')
plt.title('Distribución de Libros por Categoría')
plt.xlabel('Número de Libros')
plt.ylabel('Categoría')
plt.show()
```
